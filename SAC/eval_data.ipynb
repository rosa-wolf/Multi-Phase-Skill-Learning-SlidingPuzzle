```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_stable = "/home/rosa/Documents/Uni/Masterarbeit/SEADS_SlidingPuzzle/SAC/checkpoints/skill_conditioned_2x2_num_skills8_neg_distFalse_movementFalse_reward_on_changeTrue_sparseTrue/evaluations.npz"
path_pran = "/home/rosa/Documents/Uni/Masterarbeit/pytorch-soft-actor-critic/2x2_sparse_comp_500.npz" 
```

```python
data_stable = np.load(path_stable, mmap_mode='r')
for k in data_stable.files:
    print(k, data_stable[k])
    
data_pran = np.load(path_pran, mmap_mode='r')
for k in data_pran.files:
    print(k, data_pran[k])
```

```python
mean_reward_stable = np.sum(data_stable["results"], axis=1) / np.sum(data_stable["ep_lengths"], axis=1)
mean_reward_pran = np.sum(data_pran["results"], axis=1) / np.sum(data_pran["ep_lengths"], axis=1)

plt.figure()

plt.plot(data_stable["timesteps"], mean_reward_stable, label="Stable Baselines3")
plt.plot(data_pran["timesteps"], mean_reward_pran, label="Tandon")


plt.xlabel("environment steps")
plt.ylabel("mean reward over 5 evaluation episodes")
plt.legend()

plt.show()
```
